```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Step 1: Load dataset
df = pd.read_csv(r"G:\students.csv")
print("üìÑ Original Data:\n", df)
```

```python
# Step 2: Check for missing values
print("\n‚ùì Missing Values:\n", df.isnull().sum())
```

```python
# Step 3: Handle missing values
df['Age'].fillna(df['Age'].mean(), inplace=True)
df['Marks'].fillna(df['Marks'].mean(), inplace=True)
df['Attendance'].fillna(df['Attendance'].mean(), inplace=True)

print("\n‚úÖ After Handling Missing Values:\n", df)
```

```python
# Step 4: Basic statistics
print("\nüìä Summary Statistics:\n", df.describe())
```

```python

# Step 5: Visualization

# Histogram of Marks
plt.figure(figsize=(6,4))
sns.histplot(df['Marks'], kde=True, color='skyblue')
plt.title("Distribution of Marks")
plt.xlabel("Marks")
plt.ylabel("Frequency")
plt.show()
```

```python
# Average Marks by Department
plt.figure(figsize=(6,4))
sns.barplot(x='Department', y='Marks', data=df)
plt.title("Average Marks by Department")
plt.ylabel("Average Marks")
plt.show()
```

```python
# Gender-wise Attendance
plt.figure(figsize=(6,4))
sns.boxplot(x='Gender', y='Attendance', data=df)
plt.title("Attendance by Gender")
plt.show()
```

```python
# Step 6: Save Clean Data
df.to_csv("students_cleaned.csv", index=False)
print("‚úÖ Cleaned data saved as 'students_cleaned.csv'")
```
